```python
from constants import BIP_EVENTS, SEASON_DATES
import pandas as pd
import concurrent.futures

# Assuming df_final is already defined

# Initialize a list to store the results
ev90_list = []

# Define the function to fetch data and calculate ev90 for each player
def calculate_ev90(player_id):
    # Get the Statcast data for the player (you can adjust the date range as needed)
    player_data = statcast_batter(SEASON_DATES[season]['REG_START'],SEASON_DATES[season]['REG_END'], player_id)
    
    # Filter the data to only include balls in play
    player_data['ball_in_play'] = player_data['events'].isin(BIP_EVENTS)
    
    # Calculate the 90th percentile exit velocity for balls in play
    ev90 = player_data.loc[player_data['ball_in_play'], 'launch_speed'].quantile(0.9)
    return ev90

# Use ThreadPoolExecutor to fetch data for all players concurrently
with concurrent.futures.ThreadPoolExecutor() as executor:
    # Map the player IDs to the executor, calculate ev90 for each player
    ev90_list = list(executor.map(calculate_ev90, df['player_id']))

# Add the calculated ev90 values as a new column to the original DataFrame
df['ev90'] = ev90_list

# Print the updated DataFrame with ev90 column
print(df[['last_name, first_name','player_id', 'ev90']].head())
```
